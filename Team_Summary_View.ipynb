```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# set the graphs to show in the jupyter notebook
%matplotlib inline

# seaborn white
sns.set_style("white")

pd.set_option('display.line_width', 5000) 
pd.set_option('display.max_columns', 60) 

team = list(Team.objects.all().values('name','o_points'))
df = pd.DataFrame.from_dict(team)
```

```python
df.head()
```

```python
df['o_points'].plot(kind='bar')

plt.xlabel('name')
plt.ylabel('o_points')
plt.title('Offensive Fantasy Points by Team - 2017')

plt.show()
```

```python
o_points_hist = sns.distplot(df.o_points)
o_points_hist.set(xlabel = 'Value', ylabel = 'Frequency', title = 'Offensive Points Histogram')
sns.despine()
```
