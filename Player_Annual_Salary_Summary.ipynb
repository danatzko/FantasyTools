```python
from stats.models import GameStats,Player
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# set the graphs to show in the jupyter notebook
%matplotlib inline

# seaborn white
sns.set_style("white")

pd.set_option('display.line_width', 5000) 
pd.set_option('display.max_columns', 60) 

player = Player.objects.filter(name='Wilson, Russell')
games = list(GameStats.objects.filter(Player=player).values())
df = pd.DataFrame.from_dict(games)
```

```python
df.head()
```

```python
russell_salary = sns.distplot(df.salary)
russell_salary.set(xlabel='Value',label='Frequency', title='Russel Wilson 2017 Salary Range')
sns.despine()
```

```python
russell_points = sns.distplot(df.points)
russell_points.set(xlabel='Value',label='Frequency', title='Russel Wilson 2017 Point Range')
sns.despine()
```

```python
sns.jointplot(x="salary", y="points", data=df)
```

```python
sns.pairplot(df)
```

```python
sns.stripplot(x="points",y="salary", data=df, jitter= True)
```

```python
sns.boxplot(x="salary",y="points",hue='Away_id',data=df)
```
